```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon, Circle

def draw_canvas():
    # Create a 2x2 canvas
    canvas = np.zeros((2, 2, 3))

    # Plot the canvas
    plt.imshow(canvas, interpolation='nearest')
    plt.xticks([])  # Remove x-axis ticks
    plt.yticks([])  # Remove y-axis ticks

def color_region(canvas, region, color):
    # Color the specified region of the canvas
    if(color=="sarı"):
      canvas[region]  = (1,1,0)
    elif(color=="yeşil"):
      canvas[region]  = (0,1,0)
    elif(color=="kırmızı"):
      canvas[region]  = (1,0,0)
    elif(color=="siyah"):
      canvas[region]  = (0,0,0)

    # Plot the updated canvas
    plt.imshow(canvas, interpolation='nearest')
    plt.xticks([])  # Remove x-axis ticks
    plt.yticks([])  # Remove y-axis ticks

def draw_square(canvas, center, size, color):
    if(color=="sarı"):
      color = (1,1,0)
    if(color=="yeşil"):
      color = (0,1,0)
    if(color=="kırmızı"):
      color = (1,0,0)
    if(color=="siyah"):
      color = (0,0,0)
    # Draw a square with center coordinates and size in the canvas
    x_center, y_center = center
    x_size, y_size = size
    square = Rectangle((x_center - x_size / 2, y_center - y_size / 2), x_size, y_size, edgecolor=color, facecolor=color)
    plt.gca().add_patch(square)
    plt.imshow(canvas, interpolation='nearest')
    plt.xticks([])
    plt.yticks([])

def draw_triangle(canvas, center, size, color):
    if(color=="sarı"):
      color = (1,1,0)
    if(color=="yeşil"):
      color = (0,1,0)
    if(color=="kırmızı"):
      color = (1,0,0)
    if(color=="siyah"):
      color = (0,0,0)
    # Draw a triangle with center coordinates and size in the canvas
    x_center, y_center = center
    x_size, y_size = size
    triangle = Polygon([(x_center - x_size / 2, y_center - y_size / 2),
                        (x_center + x_size / 2, y_center - y_size / 2),
                        (x_center, y_center + y_size / 2)], edgecolor=color, facecolor=color)
    plt.gca().add_patch(triangle)
    plt.imshow(canvas, interpolation='nearest')
    plt.xticks([])
    plt.yticks([])

def draw_circle(canvas, center, radius, color):
    if(color=="sarı"):
      color = (1,1,0)
    if(color=="yeşil"):
      color = (0,1,0)
    if(color=="kırmızı"):
      color = (1,0,0)
    if(color=="siyah"):
      color = (0,0,0)
    # Draw a circle with center coordinates and radius in the canvas
    x_center, y_center = center
    circle = Circle((x_center, y_center), radius, edgecolor=color, facecolor=color)
    plt.gca().add_patch(circle)
    plt.imshow(canvas, interpolation='nearest')
    plt.xticks([])
    plt.yticks([])


# işler
#       A
#================
# ÜÇGEN KALIP
# YEŞİLE BOYA

shapes = ['kare', 'üçgen', 'yuvarlak']
colors = ['kırmızı', 'sarı', 'yeşil']

def print_shape_color(text):
    words = text.split(" ")
    color = None
    shape = None
    state = 0
    for word in words:
        gorevler=[]
        if word in colors:
            if(state==1):
              state=0
              gorevler.add(color)
            else:
              color = word
              state = 1
        elif word in shapes:
            shape = word
            if color:
                if shape == 'kare':
                    gorevler.add(f"kare {color}")
                elif shape == 'üçgen':
                    gorevler.add(f"üçgen {color}")
                elif shape == 'yuvarlak':
                    gorevler.add(f"yuvarlak {color}")
                color = "siyah"
                state=0
        else:
            print(f"Geçersiz kelime: {word}")
    return gorevler

def işle(bolge,gorev):
  gorev = gorev.split(" ")
  if(len(gorev)==1):
    coordinates_dict = {
    'A': (1, 0),
    'B': (1, 1),
    'C': (0, 0),
    'D': (0, 1)}
    color_region(canvas, tuple([round(i) for i in coordinates_dict[bolge]]), gorev[0])
  elif(len(gorev)==2):
    coordinates_dict = {
    'A': (0.01, 1),
    'B': (1.01, 1),
    'C': (0.01, 0.01),
    'D': (1.01, 0.01)}

    if(gorev[0]=="kare"):
      draw_square(canvas, coordinates_dict[bolge], (0.4, 0.4), gorev[1])  # Draw a smaller blue square centered at (0.5, 0.5)
    elif(gorev[0]=="üçgen"):
      draw_triangle(canvas, coordinates_dict[bolge], (0.4, 0.4), gorev[1])  # Draw a smaller green triangle centered at (1.5, 0.5)
    elif(gorev[0]=="yuvarlak"):
      draw_circle(canvas, coordinates_dict[bolge], 0.2, gorev[1])  # Draw a smaller purple circle centered at (1.5, 1.5)
    else:
      print("Hata Şekil")
    gorev[1]
  else:
    print("Hata")

# A_Reg_His = ["ucgen","yesil"]

canvas = np.zeros((2, 2, 3))
draw_canvas()

işle("A","üçgen yeşil")
işle("B","yuvarlak yeşil")
işle("D","kare kırmızı")
işle("C","sarı")

plt.show()
```

```python
import random

def birey_olustur(gen_sayisi):
    return [random.randint(0, 1) for _ in range(gen_sayisi)]

def popülasyon_olustur(birey_sayisi, gen_sayisi):
    return [birey_olustur(gen_sayisi) for _ in range(birey_sayisi)]

def uygunluk_hesapla(birey):
    return sum(birey)

def seçim(popülasyon, uygunluklar, seçim_limiti):
    return random.choices(popülasyon, weights=uygunluklar, k=seçim_limiti)

def çaprazlama(ebeveyn1, ebeveyn2):
    nokta = random.randint(1, len(ebeveyn1) - 1)
    çocuk1 = ebeveyn1[:nokta] + ebeveyn2[nokta:]
    çocuk2 = ebeveyn2[:nokta] + ebeveyn1[nokta:]
    return çocuk1, çocuk2

def mutasyon(birey, mutasyon_oranı):
    for i in range(len(birey)):
        if random.random() < mutasyon_oranı:
            birey[i] = 1 - birey[i]
    return birey

def genetik_algoritma(popülasyon, uygunluk_fonksiyonu, seçim_limiti, çaprazlama_oranı, mutasyon_oranı, nesil_limiti):
    uygunluklar = [uygunluk_fonksiyonu(birey) for birey in popülasyon]
    en_iyi_birey = max(popülasyon, key=uygunluk_fonksiyonu)
    for nesil in range(nesil_limiti):
        yeni_popülasyon = []
        for _ in range(len(popülasyon)):
            ebeveynler = seçim(popülasyon, uygunluklar, seçim_limiti)
            çocuk1, çocuk2 = çaprazlama(*ebeveynler)
            çocuk1 = mutasyon(çocuk1, mutasyon_oranı)
            çocuk2 = mutasyon(çocuk2, mutasyon_oranı)
            yeni_popülasyon.extend([çocuk1, çocuk2])
        popülasyon = yeni_popülasyon
        uygunluklar = [uygunluk_fonksiyonu(birey) for birey in popülasyon]
        en_iyi_birey_yeni = max(popülasyon, key=uygunluk_fonksiyonu)
        if uygunluk_fonksiyonu(en_iyi_birey_yeni) > uygunluk_fonksiyonu(en_iyi_birey):
            en_iyi_birey = en_iyi_birey_yeni
    return en_iyi_birey, uygunluk_fonksiyonu(en_iyi_birey)

# Örnek kullanım
gen_sayisi = 10
birey_sayisi = 20
seçim_limiti = 10
çaprazlama_oranı = 0.8
mutasyon_oranı = 0.01
nesil_limiti = 100

popülasyon = popülasyon_olustur(birey_sayisi, gen_sayisi)
en_iyi_birey, en_iyi_uygunluk = genetik_algoritma(popülasyon, uygunluk_hesapla, seçim_limiti, çaprazlama_oranı, mutasyon_oranı, nesil_limiti)
print(f"En iyi birey: {en_iyi_birey}, En iyi uygunluk: {en_iyi_uygunluk}")
```
