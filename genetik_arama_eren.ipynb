```python
#@title Hat İstasyon Görev Sınıfı Tanımları

import pandas as pd
import numpy as np
import requests

#En küçük görev
class Task:
    def __init__(self, job, region, duration):
        self.job = job
        self.region = region
        self.duration = duration
    def disp(self):
        return pd.DataFrame({
            'Job': [self.job],
            'Region': [self.region],
            'Duration': [self.duration]
        })
    def toList(self):
        return [self.job,self.region,self.duration]



class Station:
    def __init__(self, tasks=None):
        self.tasks = [Task() for _ in range(tasks)]  if tasks is not None else []

    def add_task(self, task):
        self.tasks.append(task)

    def get_average_time(self):
        if not self.tasks:
            return 0

        total_time = sum(task.duration for task in self.tasks)
        average_time = total_time / len(self.tasks)
        return average_time

    def get_total_time(self):
        if not self.tasks:
            return 0

        return sum(task.duration for task in self.tasks)

    def get_tasks(self):
        return self.tasks

class Line:
    def __init__(self, stations=None):
        self.stations = [Station() for _ in range(stations)] if stations is not None else []

    def get_station_average_time(self):
        if not self.stations:
            return 0

        #station her birinin ayrı ayrı ortalamasını ekrana yazdırır
        for i in range(len(self.stations)):
          print(i," Numaralı İstasyonun Ortalama Zamanı: ",self.stations[i].get_average_time())

    def total_average_time(self):
        if not self.stations:
            return 0

        listtime=[ self.stations[i].get_average_time() for i in range(len(self.stations)) ]
        return sum(listtime)/len(listtime)

    def display_stations(self):
            if not self.stations:
                return 0

            for idx, station in enumerate(self.stations):
                tasks_info = [
                    [task.job, task.region, task.duration]
                    for task in station.get_tasks()
                ]
                total_time = station.get_total_time()
                print(f"{idx} Numaralı İstasyonun {total_time} zamandaki İşleri: {tasks_info}")

    def add_station(self, station):
        self.stations.append(station)

    def get_stations(self):
        return self.stations

    def toList(self):
        return [[ j.toList() for j in i.get_tasks()] for i in self.stations]

```

```python
#@title Genetik Algoritma Şablonu

import random

#def birey_olustur(gen_sayisi):
#    return [random.randint(0, 1) for _ in range(gen_sayisi)]

# def popülasyon_olustur(birey_sayisi, gen_sayisi):
#     return [birey_olustur(gen_sayisi) for _ in range(birey_sayisi)]

def uygunluk_hesapla(birey):
    return sum(birey)

def seçim(popülasyon, uygunluklar, seçim_limiti):
    return random.choices(popülasyon, weights=uygunluklar, k=seçim_limiti)

def çaprazlama(ebeveyn1, ebeveyn2):
    nokta = random.randint(1, len(ebeveyn1) - 1)
    çocuk1 = ebeveyn1[:nokta] + ebeveyn2[nokta:]
    çocuk2 = ebeveyn2[:nokta] + ebeveyn1[nokta:]
    return çocuk1, çocuk2

def mutasyon(birey, mutasyon_oranı):
    for i in range(len(birey)):
        if random.random() < mutasyon_oranı:
            birey[i] = 1 - birey[i]
    return birey

def genetik_algoritma(popülasyon, uygunluk_fonksiyonu, seçim_limiti, çaprazlama_oranı, mutasyon_oranı, nesil_limiti):
    uygunluklar = [uygunluk_fonksiyonu(birey) for birey in popülasyon]
    en_iyi_birey = max(popülasyon, key=uygunluk_fonksiyonu)
    for nesil in range(nesil_limiti):
        yeni_popülasyon = []
        for _ in range(len(popülasyon)):
            ebeveynler = seçim(popülasyon, uygunluklar, seçim_limiti)
            çocuk1, çocuk2 = çaprazlama(*ebeveynler)
            çocuk1 = mutasyon(çocuk1, mutasyon_oranı)
            çocuk2 = mutasyon(çocuk2, mutasyon_oranı)
            yeni_popülasyon.extend([çocuk1, çocuk2])
        popülasyon = yeni_popülasyon
        uygunluklar = [uygunluk_fonksiyonu(birey) for birey in popülasyon]
        en_iyi_birey_yeni = max(popülasyon, key=uygunluk_fonksiyonu)
        if uygunluk_fonksiyonu(en_iyi_birey_yeni) > uygunluk_fonksiyonu(en_iyi_birey):
            en_iyi_birey = en_iyi_birey_yeni
    return en_iyi_birey, uygunluk_fonksiyonu(en_iyi_birey)

# Örnek kullanım
gen_sayisi = 10
birey_sayisi = 20
seçim_limiti = 10
çaprazlama_oranı = 0.8
mutasyon_oranı = 0.01
nesil_limiti = 100

popülasyon = popülasyon_olustur(birey_sayisi, gen_sayisi)
en_iyi_birey, en_iyi_uygunluk = genetik_algoritma(popülasyon, uygunluk_hesapla, seçim_limiti, çaprazlama_oranı, mutasyon_oranı, nesil_limiti)
print(f"En iyi birey: {en_iyi_birey}, En iyi uygunluk: {en_iyi_uygunluk}")
```
